Exemple de modèle et résolution du problème de Fabrication du ciment

```julia
#importer les packages utiles, le manager de package Pkg etant un package
import Pkg; Pkg.add("Cbc")
Pkg.add("JuMP")
```

```julia
# fonctionne pour JuMP version 0.21.5
using Cbc
using JuMP

# data
N = 2 # nombre de ciments disponibles
c = [50,70] # prix de vente par ciment
b = [360,480] # temps de disponibilité du four et de l'atelier
A = [40 12;20 30] # temps nécessaire de calcination (four) et de broyage (atelier) par ciment

# set optimizer
model = Model(Cbc.Optimizer)

# define variables
@variable(model, ciment[1:N] >= 0)

# define objective function
@objective(model, Max, sum(c[i]*ciment[i] for i in 1:N))

# define constraints
for i in 1:length(b)
    @constraint(model, sum(A[i,j]*ciment[j] for j in 1:N) <= b[i])
end

# run optimization
optimize!(model)
```

```julia
# print solution
println("Solution obtenue:")
println("\t benefice = $(objective_value(model))")
for i in 1:N
    println("\t quantite de ciment $i = $(value(ciment[i]))")
end
```

```julia
# Cas particulier 1

# fonctionne pour JuMP version 0.21.5
using Cbc
using JuMP

# data
Nf = 2 # nombre de fluides
Ne = 3 # nombre d'entrepôts
Nd = 2 # nombre de demandes
p = [1 2 3;1 3 2] # prix de vente
s = [2.5 1 2;1 2 1] # stocks
d = [2 1;0 3] # demandes

# set optimizer
model = Model(Cbc.Optimizer)

# define variables
@variable(model, quantites[1:Nf,1:Ne] >= 0)

# define objective function
@objective(model, Min, sum(sum(p[f,e]*quantites[f,e] for f in 1:Nf) for e in 1:Ne))

# define constraints
for f in 1:Nf
    @constraint(model, sum(quantites[f,e] for e in 1:Ne) == sum(d[f,j] for j in 1:Nd))
    for e in 1:Ne
        @constraint(model, quantites[f,e] <= s[f,e])
    end
end

# run optimization
optimize!(model)
```

```julia
# print solution
println("Solution obtenue:")
println("\t coût total = $(objective_value(model))")
for f in 1:Nf
    for e in 1:Ne
        println("\t quantité de fluide $f à acheter dans l'entrepôt $e = $(value(quantites[f,e]))")
    end
end
```

```julia
# Cas particulier 3

# fonctionne pour JuMP version 0.21.5
using Cbc
using JuMP

# data
Nf = 2 # nombre de fluides
Ne = 3 # nombre d'entrepôts
Nd = 2 # nombre de demandes
p = [1 2 3;1 3 2] # prix de vente
s = [2.5 1 2;1 2 1] # stocks
d = [3 1;0 3] # demandes
exp = [1 0 0; 0 2 1] #coût expédition
somme_stocks = sum(s)

# set optimizer
model = Model(Cbc.Optimizer)

# define variables
@variable(model, quantites[1:Nf,1:Ne,1:Nd] >= 0, Int)
@variable(model, 0 <= offre[1:Nd,1:Ne] <= 1, Int)

# define objective function
@objective(model, Min, sum(sum(sum(p[f,e]*quantites[f,e,dem] for f in 1:Nf) for e in 1:Ne) for dem in 1:Nd) + sum(sum(offre[dem,e]*exp[dem,e] for dem in 1:Nd) for e in 1:Ne))

# define constraints
for f in 1:Nf
        for dem in 1:Nd
            @constraint(model, sum(quantites[f,e,dem] for e in 1:Ne) == d[f,dem])
        end
    end

for e in 1:Ne
    for dem in 1:Nd
            @constraint(model,somme_stocks*offre[dem,e]>=sum(quantites[f,e,dem] for f in 1:Nf))
    end
    for f in 1:Nf
        @constraint(model, sum(quantites[f,e,dem] for dem in 1:Nd) <= s[f,e])
    end
end

# run optimization
optimize!(model)
```

```julia
# print solution
println("Solution obtenue:")
println("\t coût total = $(objective_value(model))")
for f in 1:Nf
    for e in 1:Ne
        for dem in 1:Nd
            println("\t quantité de fluide $f à acheter dans l'entrepôt $e pour la demande $dem = $(value(quantites[f,e,dem]))")
        end
    end
end
for e in 1:Ne
    for dem in 1:Nd
        println("\t Offre de l'entrepôt $e pour la demande $dem = $(value(offre[dem,e]))")
    end
end
```
